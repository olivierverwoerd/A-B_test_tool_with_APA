# Bereken het minimal aantal gebruikers voor de A/B test
## LET OP: Deze gegevens worden mee genomen in het advies

```python
from scipy.stats import norm
import numpy as np
import math

def sample_size_calculator(base_conversion_rate, mde, power=0.8, alpha=0.05):
    sd1 = np.sqrt(2 * base_conversion_rate * (1 - base_conversion_rate))
    sd2 = np.sqrt((base_conversion_rate + mde) * (1 - (base_conversion_rate + mde)))
    min_sample_size = (norm.ppf(1-alpha/2)*sd1 + norm.ppf(power)*sd2) ** 2 / mde ** 2
    return round(min_sample_size)

# -------------PAS DIT AAN VOOR JE MINIMAAL AANTAL GEBRUIKERS-----------------
# huidige conversie in %, in het voorbeeld 10%
base_conversion_rate = 15

# minimaal verschil wat er gemeten moet worden in %, in het voorbeeld 2%
mde = 2 

# Hoeveel varianten wil je testen?
amount_of_groups = 4 

# Optioneel. hoeveel gebruikers zijn er per dag?
users_per_day = 1000
sample_size = sample_size_calculator(base_conversion_rate /100, mde /100)
# -----------------------------------------------------------------------------

print(f"Er zijn op zijn minst: {sample_size} gebruikers nodig per group")
print(f"In totaal dus: {sample_size*amount_of_groups} gebruikers")
if users_per_day:
  print(f"De test duurt ongeveer: {math.floor(((sample_size*amount_of_groups)/users_per_day)/7)} weken en {math.ceil(((sample_size*amount_of_groups)/users_per_day)%7)} dagen")
```

## Verwerken data
Als de data gewenst klopt kan je het blok hieronder uitvoeren

```python


# Calculate conversion rates for each group
control = data[0]
control_conversion_rate = control["goal"] / control["start"]
control_standard_error = np.sqrt(control_conversion_rate * (1 - control_conversion_rate) / control["start"])
control_deviation = np.sqrt(control["start"] * control_conversion_rate * (1 - control_conversion_rate))

conversion_rates = []
standard_errors = []
standard_deviations = []
relative_uplifts = []
chance_outperform = []
degrees_of_freedom = []
p_values = []
z_values = []

for group in data[1:]:
    conversion_rate = group["goal"] / group["start"]
    standard_error = np.sqrt(conversion_rate * (1 - conversion_rate) / group["start"])
    std_dev = np.sqrt(group["start"] * conversion_rate * (1 - conversion_rate))
    conversion_rates.append(conversion_rate)
    standard_errors.append(standard_error)
    standard_deviations.append(std_dev)
    

    uplift = (conversion_rate - control_conversion_rate) / control_conversion_rate
    relative_uplifts.append(uplift)

    # Calculate percentage chance of outperforming the control group
    n_control = control["start"]
    n_group = group["start"]
    p_control = control_conversion_rate
    p_group = conversion_rate

    #calc df, z-score, p-value
    degrees_of_freedom.append(n_group + n_control - 2)
    total_conversions = n_control + n_group
    expected_control = total_conversions * (control["start"] / (control["start"] + group["start"]))
    expected_group = total_conversions * (group["start"] / (control["start"] + group["start"]))
    z_score = (p_group - p_control) / np.sqrt((p_control * (1 - p_control) / n_control) + (p_group * (1 - p_group) / n_group))
    p_value = 1 - stats.norm.cdf(abs(z_score))
    chance = (1 - p_value) * 100

    p_values.append(p_value)
    z_values.append(z_score)
    chance_outperform.append(chance)

# Print conversion rates, relative uplifts, standard errors, and chances of outperforming the control group
print("----------Advies----------")
nothing = True
for i, rate in enumerate(conversion_rates):
        
        
    
    if (p_values[i] < 0.05 and chance_outperform[i] > 55):
        if (degrees_of_freedom[i] + 2 > sample_size * 2):
            print(f"Gebruik de elementen van {data[i+1]['name']}")
        else:
            print(f"Je toets heeft nog niet het benodigde aantal gebruikers, maar de elementen van {data[i+1]['name']} lijken positief")
        nothing = False
    elif (p_values[i] < 0.05 and chance_outperform[i] < 45):
        
        
        if (degrees_of_freedom[i] + 2 > sample_size * 2):
            print(f"Gebruik niet de elementen van {data[i+1]['name']}")
        else:
            print(f"Je toets heeft nog niet het benodigde aantal gebruikers, maar de elementen van {data[i+1]['name']} lijken negatief")
        nothing = False
        

if nothing:
    print("Doe niks. Wacht de test af tot een beter resultaat.")
print()


print("----------APA----------")
print("APA richtlijn")
print("1. De meeste symbolen en afkortingen die geen Griekse letters zijn moeten cursief:(df p N n t z F SD SE M OR)")
print("2. Gebruik een hoofdletter N voor het getal in de totale steekproef (N = 45) en een kleine letter n voor een fractie van de steekproef (n = 20).")
print("")
print("Resultaten apa tekst:")
for i, rate in enumerate(conversion_rates):
    if (p_values[i] < 0.001 and chance_outperform[i] > 55):
        print(f"Het verschil tussen de groep {data[i+1]['name']} (M = {rate:.2f}; SD = {standard_deviations[i]:.3f}) en de controle groep (M = {control_conversion_rate:.2f}; SD = {control_deviation:.3f}) is zeer significant (z = {z_values[i]:.2f}; p < .001).")
    elif (p_values[i] < 0.05 and chance_outperform[i] > 55):
        print(f"Het verschil tussen de groep {data[i+1]['name']} (M = {rate:.2f}; SD = {standard_deviations[i]:.3f}) en de controle groep (M = {control_conversion_rate:.2f}; SD = {control_deviation:.3f}) is significant (z = {z_values[i]:.2f}; p = {remove_leading_zero(p_values[i])}).")
    else:
        print(f"Het verschil tussen de groep {data[i+1]['name']} (M = {rate:.2f}; SD = {standard_deviations[i]:.3f}) en de controle groep (M = {control_conversion_rate:.2f}; SD = {control_deviation:.3f})is niet significant (z = {z_values[i]:.2f}; p = {remove_leading_zero(p_values[i])}).")

        
print()
    
```

## Controle voor normaal verdeling
Om te controleren dat dit daadwerkelijk een niet normaal verdeelde data bevat wordt hier onder een Shapiro-Wilk Test en een Q-Q plot gemaakt.
Als de Shapiro-Wilk Test een p waarde heeft van onder de 0.05 dan is de data niet normaal verdeeld. 
Als de lijn en de data in het Q-Q plot niet overeen komen is de data niet normaal verdeeld.

```python

for key in data:
    print("Shapiro-Wilk Test for Control: ", stats.shapiro(df[df['Group'] == key]['Activities']))
    plt.figure(figsize=(6,4))
    stats.probplot(data[key], dist="norm", plot=plt)
    plt.title('Q-Q plot for '+ key)
    plt.show()
```

## Bootstrapped-t-toets
hieronder doen we bootstrapped-t-toets omdat de data niet normaal verdeeld is.

```python
def remove_leading_zero(number):
    if number < 0.001:
        return 0.001
    elif number < 1:
        return str(float(f"{number:.3f}"))[1:]
    else:
        return str(float(f"{number:.3f}"))

def bootstrap_ttest(group_data, control_data, n_replicas=100000):
    diff_mean=np.mean(group_data)-np.mean(control_data)
    
    # combine the data
    combined_data = np.concatenate((group_data,control_data))

    n1 = len(group_data)
    n2 = len(control_data)
    diffs = []
    for _ in range(n_replicas):
        permuted_data = np.random.permutation(combined_data)
        new_group = permuted_data[:n1]
        new_control = permuted_data[n1:]
        diffs.append(np.mean(new_group) - np.mean(new_control))
    p_value = np.sum(diffs >= diff_mean) / n_replicas
    return diff_mean, p_value


groups = []
for group in data:
    groups.append(group)


control_data = df[df['Group'] == 'Control']['Activities']

table = PrettyTable()
table.field_names = ['Groep', 'M', 'SD','SE', 'df', 't', 'p', 'd', 'CI Lower', 'CI Upper']

x = np.linspace(np.min(control_data), np.max(control_data), 100)
plt.figure(figsize=(12, 6))

for group in groups:
    group_data = df[df['Group'] == group]['Activities']
    
    # Calculate mean and standard deviation
    mean = np.mean(group_data)
    std_dev = np.std(group_data)
    
    # Confidence Interval calculation
    confidence_level = 0.95
    degrees_freedom = group_data.size - 1
    critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_freedom)
    standard_error = std_dev / np.sqrt(group_data.size)

    ci_lower = mean - critical_value * standard_error
    ci_upper = mean + critical_value * standard_error
    
    if group == 'Control':
        table.add_row([group, "{:.2f}".format(mean), "{:.2f}".format(std_dev), "{:.3f}".format(standard_error), "-", "-","-","-", "{:.2f}".format(ci_lower), "{:.2f}".format(ci_upper)])
        plt.plot(x, stats.norm.pdf(x, mean, std_dev), label=group)
    else: 
        # t-test
        t_stat, p_val = bootstrap_ttest(group_data, control_data)
        
        
        dfreedom = len(group_data) + len(control_data) - 2
        # Calculate cohens d
        cohens_d = (np.mean(group_data) - np.mean(control_data)) / np.sqrt((np.var(group_data) + np.var(control_data)) / 2)
        plt.plot(x, stats.norm.pdf(x, mean, std_dev), label=group)
        table.add_row([
            group, 
            "{:.2f}".format(mean), 
            "{:.2f}".format(std_dev), 
            "{:.3f}".format(standard_error), 
            dfreedom, 
            "{:.3f}".format(t_stat),
            remove_leading_zero(p_val),
            "{:.3f}".format(cohens_d),
            "{:.2f}".format(ci_lower),
            "{:.2f}".format(ci_upper)
        ])

print(table) 

plt.title('Dichtheid van de activiteiten per gebruiker per groep') 
plt.xlabel('Activiteiten per gebruiker')
plt.ylabel('Dichtheid')
plt.legend()
plt.grid(True)
plt.show()
```

## Bootstrapped-t-toets plot
hieronder doen we nog een keer een bootstrapped-t-toets met een hogere steekproef grote voor een betere grafiek van de data in verhouding met de controle groep.

```python
def draw_bs_pairs(group_data, control_data, size=1):
    inds1 = np.arange(len(group_data))
    inds2 = np.arange(len(control_data))
    bs_group_means = np.empty(size)
    bs_control_means = np.empty(size)
    for i in range(size):
        bs_inds1 = np.random.choice(inds1, size=len(inds1))
        bs_inds2 = np.random.choice(inds2, size=len(inds2))
        bs_group_means[i] = group_data[bs_inds1].mean()
        bs_control_means[i] = control_data[bs_inds2].mean()
    return bs_group_means, bs_control_means

plt.figure(figsize=(12, 6))
for group in groups[1:]:
    group_data = df[df['Group'] == group]['Activities'].values
    control_data = df[df['Group'] == 'Control']['Activities'].values
    bs_group_means, bs_control_means = draw_bs_pairs(group_data, control_data, size=10000)
    mean_diffs = bs_group_means - bs_control_means
    
    # plot histogram of mean differences
    _ = plt.hist(mean_diffs, bins=50, alpha=0.5, label=group)
    _ = plt.axvline(x=np.mean(mean_diffs), color='r', linestyle='dashed', linewidth=2)

plt.title('Bootstrapverdeling van gemiddelde verschillen tussen de onboardingsvariant en de controle')
plt.xlabel('Gemiddeld verschil ten opzichte van de controle groep')
plt.ylabel('Frequentie')
plt.legend()
plt.grid(True)
plt.show()
```
